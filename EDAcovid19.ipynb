# This notebook contains a demo of basic Exploratory Data Analysis on covid-19 data set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid=pd.read_csv('/content/Latest Covid-19 India Status.csv')
covid.head()
```

```python
covid.shape
```

```python
covid.columns
```

```python
covid.isna().sum()
```

```python
covid.describe()
```

```python
cv=covid.copy()
cv.head()
```

```python
cv=cv.sort_values(by=['Total Cases'],ascending=False)
cv.reset_index(drop=True,inplace=True)
```

```python
cv=cv.sort_values(by='Total Cases')
```

```python
sns.set()
plt.figure(figsize=(10,5))
plt.title('STATE vs TOTAL CASES')
sns.barplot(cv['State/UTs'],cv['Total Cases'],palette='Blues_d')
plt.xticks(rotation=90)
plt.xlabel('States/UTs')
plt.ylabel('Total Cases In Million')
plt.show()
```

```python
plt.pie(cv["Active"].loc[30:],labels=cv["State/UTs"].loc[30:], autopct='%.1f%%',startangle=0)
plt.axis("equal")
plt.title("Top 6 states with highest active cases.")
plt.show()
```

```python
cv=cv.sort_values(by=['Discharged'])
cv.reset_index(drop=True,inplace=True)
cv.head()
```

```python
sns.set()
plt.figure(figsize=(10,5))
plt.title('STATE vs TOTAL CASES')
sns.barplot(cv['State/UTs'],cv['Discharged'],palette='Purples_d')
plt.xticks(rotation=90)
plt.xlabel('States/UTs')
plt.ylabel('Total Cases In Million')
plt.show()
```

```python
sns.set()
plt.figure(figsize=(10,5))
plt.title('STATE vs DEATHS')
sns.barplot(cv['State/UTs'],cv['Deaths'],palette='Reds_d')
plt.xticks(rotation=90)
plt.xlabel('States/UTs')
plt.ylabel('Total Deaths')
plt.show()
```

```python
covid=covid.sort_values(by= 'Deaths')
```

```python
plt.pie(covid["Deaths"].loc[30:],labels=covid["State/UTs"].loc[30:], autopct='%.1f%%',startangle=40)
plt.axis("equal")
plt.title("Top 5 states with highest deaths.")
plt.show()
```

# Relation between 2 varibales

```python
sns.set()

sns.jointplot(x=cv['Total Cases'], y=cv['Discharged'],kind="reg", color='Purple')

plt.show()
```

```python
sns.set()

sns.jointplot(x=cv['Total Cases'], y=cv['Deaths'],kind="reg", color='Green')

plt.show()
```
